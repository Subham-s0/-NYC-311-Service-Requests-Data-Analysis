Changing the datatype of 'Created date' and 'Closed Date' column to datetime

```python
df_service_request['Created Date'] = pd.to_datetime(df_service_request['Created Date'], format='mixed',dayfirst=False)
#to  it treats the first number as the month if the montha anda date are same
df_service_request['Closed Date'] = pd.to_datetime(df_service_request['Closed Date'], format='mixed',dayfirst=False)
```

Creating a new column 'Request_Closing_Time' time

```python
df_service_request['Request_Closing_Time'] = (df_service_request['Closed Date'] - df_service_request['Created Date'])
df_service_request['Request_Closing_Time'] = df_service_request['Request_Closing_Time'].dt.total_seconds() / 3600# converting time to hours
```

```python
new_df_dates = df_service_request[['Created Date', 'Closed Date', 'Request_Closing_Time']]
new_df_dates
```

Dropping irrelevant Columns from the data set

```python
dropped_column = [
    'Agency Name', 'Incident Address', 'Street Name', 'Cross Street 1', 'Cross Street 2',
    'Intersection Street 1', 'Intersection Street 2', 'Address Type', 'Park Facility Name',
    'Park Borough', 'School Name', 'School Number', 'School Region', 'School Code',
    'School Phone Number', 'School Address', 'School City', 'School State', 'School Zip',
    'School Not Found', 'School or Citywide Complaint', 'Vehicle Type', 'Taxi Company Borough',
    'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp',
    'Bridge Highway Segment', 'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name',
    'Landmark', 'X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Due Date',
    'Resolution Action Updated Date', 'Community Board', 'Facility Type', 'Location'
]
df_service_request.drop(columns=dropped_column,inplace=True)# to modify the existing dataset
df_service_request.shape
```

```python
df_service_request.isnull().sum()
```

Removing the NaN missing values

```python
df_service_request = df_service_request.dropna()
df_service_request.shape
```

Finding unique values from all the columns

```python
import pandas as pd
unique_list = []
for column in df_service_request.columns:
    unique_values = df_service_request[column].unique()#give unique values in numpy array
    total_unique = df_service_request[column].nunique()# =give the count of unique values
    if len(unique_values) > 2:
        display_values = [unique_values[0], "...", unique_values[-1]]
    else:
        display_values = unique_values.tolist()
    unique_list.append({
        "Column Name": column,
        "Total Unique Values": total_unique,
        "Unique Values": display_values
    })
unique_df = pd.DataFrame(unique_list)

unique_df
```

summary statistics of the data

```python
numeric_df = df_service_request.select_dtypes(include=[np.number])#inclues the columns having  only numeric data
statistics_df= numeric_df.agg(['sum', 'mean', 'std', 'skew', 'kurt'])#.agg runs all the method inside its parameter on the df
statistics_df.index=('Sum','Mean','Standard Deviation','skewness','kurtosis')
statistics_df
```

Top 10 Complaints types

```python
import matplotlib.pyplot as plt
#sorting the complaint type in descentding order by grouping them with the count of complaint type 
top_complaints = df_service_request['Complaint Type'].value_counts().head(10).sort_values(ascending=False)

plt.figure(figsize=(12, 8))
current_axes = plt.gca()
bars = plt.bar(top_complaints.index, top_complaints.values,color='skyblue')

# Adding  the number of complaints on top of each bar
for bar in bars:
    height = bar.get_height()
    x_position = bar.get_x() + bar.get_width() / 2
    y_position = height + 0.02 * max(top_complaints.values)
    label = "{:,}".format(int(height))
    current_axes.text(x_position, y_position, label,ha='center', va='bottom')

# Add labels and formatting
plt.title('Top 10 Most Frequent Complaint Types')
plt.ylabel('Number of Complaints')
plt.xlabel('Complaint Type')
plt.xticks(rotation=45, ha='right')
current_axes.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

Average Closing Time for Top 10 Most Frequent Complaint Types

```python
import matplotlib.pyplot as plt
#top complaints
top_complaints = df_service_request['Complaint Type'].value_counts().head(10).index
median_closing_times = df_service_request[df_service_request['Complaint Type'].isin(top_complaints)] \
    .groupby('Complaint Type')['Request_Closing_Time'].median().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
current_axes = plt.gca()
bars = plt.bar(median_closing_times.index, median_closing_times.values, color='skyblue')
for bar in bars:
    height = bar.get_height()
    x_position = bar.get_x() + bar.get_width() / 2
    y_position = height + 0.02 * max(median_closing_times.values)
    label = "{:.1f} hours".format(height)
    current_axes.text(x_position, y_position, label, ha='center', va='bottom')

plt.title('Average Closing Time for Top 10 Most Frequent Complaint Types')
plt.ylabel('Average Closing Time (hours)')
plt.xlabel('Complaint Type')
plt.xticks(rotation=45, ha='right')
current_axes.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

Average Daily Requests per Day of the Week

```python
import pandas as pd
import matplotlib.pyplot as plt

df_new = df_service_request.copy() 
df_new['Week'] = df_new['Created Date'].dt.isocalendar().week
df_new['Year'] = df_new['Created Date'].dt.year
df_new['Day'] = df_new['Created Date'].dt.day_name()
weekly_day_counts = df_new.groupby(['Year', 'Week', 'Day']).size().reset_index(name='Count')
weekly_totals = weekly_day_counts.groupby(['Year', 'Week'])['Count'].transform('sum')
weekly_day_counts['Percent'] = (weekly_day_counts['Count'] / weekly_totals) * 100
avg_percentages = weekly_day_counts.groupby('Day')['Percent'].mean()
service_days = [ 'Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
avg_percentages = avg_percentages.reindex(service_days)
plt.figure(figsize=(8, 8))
plt.pie(avg_percentages.values, labels=avg_percentages.index, autopct='%1.1f%%', startangle=0, colors=plt.cm.Paired.colors)
plt.title('Average Daily Percentage of Weekly Service Requests')
plt.tight_layout()
plt.show()
```

Distribution of Request_Closing_time removing the outliers

```python
import matplotlib.pyplot as plt

Q1 = df_service_request['Request_Closing_Time'].quantile(0.25)
Q3 = df_service_request['Request_Closing_Time'].quantile(0.75)
Inter_quartile_range = Q3 - Q1
lower_bound = Q1 - 1.5 * Inter_quartile_range
upper_bound = Q3 + 1.5 * Inter_quartile_range
filtered_data = df_service_request[(df_service_request['Request_Closing_Time'] >= lower_bound) & (df_service_request['Request_Closing_Time'] <= upper_bound)]
min_val = int(np.floor(filtered_data['Request_Closing_Time'].min()))
max_val = int(np.ceil(filtered_data['Request_Closing_Time'].max()))
bins = np.arange(min_val, max_val + 1, 1)
plt.figure(figsize=(8, 6))
plt.hist(filtered_data['Request_Closing_Time'], bins=bins, color='#8fd3fe', edgecolor='#daf0ff', linewidth=1)
plt.xticks(bins)
plt.title('Distribution of Request Closing Time')
plt.xlabel('Request Closing Time (hours)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

Number of complaints per Borough with the complaint types

```python
import seaborn as sns
import matplotlib.pyplot as plt

heatmap = df_service_request.pivot_table(
    index='Complaint Type',
    columns='Borough',
    aggfunc='size',
    fill_value=0
)
heatmap.loc['Total Complaints'] = heatmap.sum()
plt.figure(figsize=(10, 12))
sns.heatmap(heatmap, cmap='Reds', linewidths=0.5, annot=True, fmt='g')
plt.title('Number of Complaints by Type and Borough')
plt.xlabel('Borough')
plt.ylabel('Complaint Type')
plt.tight_layout()
plt.show()
```

Average Request Closing Time by Complaint Type and Location Type

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
grouped_data = (df_service_request.groupby(['Complaint Type', 'Location Type'])['Request_Closing_Time'].median()
                .reset_index().sort_values('Request_Closing_Time', ascending=False))
pivot_data = grouped_data.pivot(index='Complaint Type',columns='Location Type',values='Request_Closing_Time')
x_labels = pivot_data.columns.tolist() 
y_labels = pivot_data.index.tolist()    

plt.figure(figsize=(12, 11))
ax = plt.gca()
im = ax.imshow(pivot_data.values, cmap='cividis', aspect='auto')
ax.set_xticks(np.arange(len(x_labels)))
ax.set_yticks(np.arange(len(y_labels)))
ax.set_xticklabels(x_labels, rotation=45, ha='right')
ax.set_yticklabels(y_labels)

colorbar = plt.colorbar(im)
colorbar.set_label('Median Closing Time (hours)', rotation=270, labelpad=20)

for i in range(len(y_labels)):
    for j in range(len(x_labels)):
        val = pivot_data.values[i, j]
        if not np.isnan(val):
            ax.text(j, i, f'{val:.1f}', ha='center', va='center', color='white', fontsize=10)
plt.title('Median Request Closing Time by Complaint Type and Location Type')
plt.xlabel('Location Type')
plt.ylabel('Complaint Type')
plt.grid(True,linestyle='--', linewidth=0.3, color='lightgrey')
plt.tight_layout()
plt.show()
```
